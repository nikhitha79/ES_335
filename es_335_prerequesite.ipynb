### Question 1

#### For matrices with dimensions $ (n,k) $ and $ (k,m) $
For each element:
    multiplication = $ k $
    addition = $ k -1 $

Total number of multiplication = $ k * m * n $

Total number of addition = $ k * (m-1) * n $

Therefore, Total number of operations = $ k * m * n + k * (m-1) * n $

### Question 3

Time complexity of sorting method is $ O(nlogn) $ and finding median by this method takes $ O(1) $ time.


```python
arr=[12,7,3,80,32,45]
sorted_arr=sorted(arr)
n=len(sorted_arr)
print('Array after sorting : ', sorted_arr)
if n%2==1:
    print('median is : ', sorted_arr[n//2])
else:
    mid=n//2
    print('median is : ', (sorted_arr[mid-1]+sorted_arr[mid])/2)

```

An alternative method can be using the quicksort algorithm which is also what numpy uses under the hood. If we do not want the entire array sorted, one can use partitioning which partitions the lesser values to the left and greater to the right of the array without actually sorting the array.

```python
def quick_sort(ar):
    if len(ar)<=1:
        return ar
    else:
        pivot_element=ar[0]
        lesser=[x for x in ar[1:] if x<=pivot_element]
        greater=[x for x in ar[1:] if x>pivot_element]
        return quick_sort(lesser) + [pivot_element] + quick_sort(greater)
ar = [3, 2, 8, 1, 4, 10, -1]
sorted_arr = quick_sort(ar)
print("Sorted Array:", sorted_arr)
```

```python
#numpy method
median=np.median(sorted_arr)
print('median is : ', median)
```

$ \frac{\partial{f}}{\partial{y}} =  x^{2} + 3y^{2}sin(x) $




#### Question 5

```python
import jax
import jax.numpy as jnp

def f(x, y):
    u = x**2 * y
    v = y**3 * jnp.sin(x)
    return u + v

# partial derivatives
grad_f = jax.grad(f, argnums=(0,1))

#  values for x and y
x_val = 2.0
y_val = 3.0

# analytical gradients
grad_x = 2 * x_val * y_val + y_val**3 * jnp.cos(x_val)
grad_y = x_val**2 + 3 * y_val**2 * jnp.sin(x_val)

# JAX  gradients
jax_grad_x, jax_grad_y = grad_f(x_val, y_val)


print("Analytical gradient with respect to x:", grad_x)
print("JAX gradient with respect to x:", jax_grad_x,'\n')
print("Analytical gradient with respect to y:", grad_y)
print("JAX gradient with respect to y:", jax_grad_y)
```

#### Question 11

```python
import numpy as np
import pandas as pd

np.random.seed(42)
random_matrix = np.random.uniform(1, 2, (20, 5))
df = pd.DataFrame(random_matrix, columns=["a", "b", "c", "d", "e"])
df['f'] = df.sum(axis=1)
df['g'] = np.where(df['f'] < 8, 'LT8', 'GT8')
# Find the number of rows where the value in column "g" is "LT8"
num_lt8_rows = df[df['g'] == 'LT8'].shape[0]

# Find the standard deviation of column "f" for rows where "g" is "LT8" and "GT8" respectively
std_lt8 = df.loc[df['g'] == 'LT8', 'f'].std()
std_gt8 = df.loc[df['g'] == 'GT8', 'f'].std()

# Display the modified DataFrame and results
print("Modified DataFrame:")
print(df)
print("\nNumber of rows where the value in column 'g' is 'LT8':", num_lt8_rows)
print("Standard deviation of column 'f' for rows where 'g' is 'LT8':", std_lt8)
print("Standard deviation of column 'f' for rows where 'g' is 'GT8':", std_gt8)
```

#### Question 13

```python
import numpy as np

def cust_argmin(arr):
    arr = np.array(arr)
    min_value = arr[0]
    min_index = 0
    
    #  find the minimum value and its index
    for i in range(0,len(arr)):
            if arr[i] < min_value:
                min_value = arr[i]
                min_index = i
    
    return min_index

arr = np.array([4, 78, 9, 9, -1, 8, 2, 3])
result_custom = cust_argmin(arr)

# np.argmin 
result_np = np.argmin(arr)

print("Custom argmin:", result_custom)
print("np.argmin result:", result_np)
```
